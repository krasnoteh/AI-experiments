```python
#автоматический сборщик картинок из браузера
#откройте word в одном окне и браузер в другом, так чтобы между ними переключатся через alt + tab
#настройте координаты поисковой строки (тоже через pyautogui)
#проверьте правильность путей сохранения ниже
#скорость сборки ~100 картинок в минуту (но зависти от настроек браузера и скорости интернета)

for c in range(10):
    print("start in ", 10 - c)
    time.sleep(1)
with open("custom_captions.txt", 'a') as cap:
    for task in tasks:
        pyautogui.click(x=225, y=172)
        pyautogui.hotkey('ctrl', 'a')
        paste(task)
        pyautogui.press('enter')
        time.sleep(5)
        pyautogui.hotkey('ctrl', 'a')
        pyautogui.hotkey('ctrl', 'c')
        pyautogui.hotkey('alt', 'tab')
        pyautogui.hotkey('ctrl', 'a')
        pyautogui.hotkey('ctrl', 'v')
        time.sleep(20)
        pyautogui.hotkey('ctrl', 's')
        time.sleep(3)
        import zipfile
        with zipfile.ZipFile("buffer.docx", 'r') as zip_ref:
            zip_ref.extractall("result")
        path = "result/word/media"
        files = [f for f in listdir(path) if isfile(join(path, f))]
        for ff in files:
            try:
                image = cv2.imread(path +"/"+ ff)
                if (image.shape[0]>= 128 and image.shape[1]>= 128):
                    image = cv2.resize(image,(128,128), interpolation = cv2.INTER_AREA)
                    name = get_random_string(16) + '.jpg'
                    cv2.imwrite('custom images/' +  name , image)
                    cap.write(name + "|" + task + "\n")
            except Exception:
                pass
        shutil.rmtree("result")
        pyautogui.hotkey('alt', 'tab')
        time.sleep(3)
```
